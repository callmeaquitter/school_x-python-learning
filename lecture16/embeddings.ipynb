```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
```

```python
with open("C:\\Users\\User\\Downloads\\Telezhka_s_yablokami.txt", 'r', encoding='cp1251') as f:
    text = f.read()
```

```python
word2vec_rus = gensim.downloader.load('word2vec-ruscorpora-300')
```

```python
tokens = word_tokenize(text)
stop_words = set(stopwords.words('russian'))
filtered_tokens = [token for token in tokens if token not in stop_words]
```

```python
import gensim

model = gensim.models.KeyedVectors.load_word2vec_format('ruscorpora_upos_skipgram_300_5_2018.vec.gz', binary=False)
```

```python
word_vectors = []
for token in filtered_tokens:
    try:
        word_vectors.append(model.wv[token])
    except KeyError:
        pass
```
