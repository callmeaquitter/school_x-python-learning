```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

with open("C:\\Users\\User\\Downloads\\Uchenik_dyavola.txt", 'r') as file:
    text = file.read()


sentences = sent_tokenize(text)
words = word_tokenize(text)


stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
filtered_words = [lemmatizer.lemmatize(w.lower()) for w in words if not w.lower() in stop_words]


fdist = nltk.FreqDist(filtered_words)
print(fdist.most_common(10))


pos_tags = nltk.pos_tag(words)
print(pos_tags)
```

```python
from natasha import (
    Segmenter,
    MorphVocab,
    NewsEmbedding,
    NewsMorphTagger,
    Doc,
    SyntaxParser
)
from graphviz import Source


with open("C:\\Users\\User\\Downloads\\Uchenik_dyavola.txt", 'r', encoding='utf-8') as f:
    text = f.read()


doc = Doc(text)


segmenter = Segmenter()
doc.segment(segmenter)


morph_vocab = MorphVocab()
emb = NewsEmbedding()
morph_tagger = NewsMorphTagger(emb)
doc.tag_morph(morph_tagger)


syntax_parser = SyntaxParser(emb)
doc.parse_syntax(syntax_parser)


dot = doc.sents[0].syntax.to_dot()


Source(dot)
```
