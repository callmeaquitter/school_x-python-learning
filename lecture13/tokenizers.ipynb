```python
import nltk
from nltk.tokenize import BlanklineTokenizer, LegalitySyllableTokenizer, LineTokenizer, MWETokenizer, NLTKWordTokenizer, PunktSentenceTokenizer, RegexpTokenizer, ReppTokenizer, SExprTokenizer, SpaceTokenizer, StanfordSegmenter, SyllableTokenizer, TabTokenizer, TextTilingTokenizer, ToktokTokenizer, TreebankWordDetokenizer, TreebankWordTokenizer, TweetTokenizer, WhitespaceTokenizer, WordPunctTokenizer

with open("C:\\Users\\User\\Downloads\\The_Ones_Who_Walk_Away_from_Omelas.txt", 'r', encoding='utf-8') as f:
    text = f.read()

tokens_blankline = BlanklineTokenizer().tokenize(text)
tokens_legality_syllable = LegalitySyllableTokenizer().tokenize(text)
tokens_line = LineTokenizer().tokenize(text)
tokens_mwe = MWETokenizer().tokenize(text.split())
tokens_nltk_word = NLTKWordTokenizer().tokenize(text)
tokens_punkt = PunktSentenceTokenizer().tokenize(text)
tokens_regex = RegexpTokenizer(r'\w+').tokenize(text)
tokens_repp = ReppTokenizer().tokenize(text)
tokens_sexpr = SExprTokenizer().tokenize(text)
tokens_space = SpaceTokenizer().tokenize(text)
tokens_stanford = StanfordSegmenter().tokenize(text)
tokens_syllable = SyllableTokenizer().tokenize(text)
tokens_tab = TabTokenizer().tokenize(text)
tokens_tiling = TextTilingTokenizer().tokenize(text)
tokens_toktok = ToktokTokenizer().tokenize(text)
tokens_treebank_detokenize = TreebankWordDetokenizer().detokenize(TreebankWordTokenizer().tokenize(text))
tokens_treebank = TreebankWordTokenizer().tokenize(text)
tokens_tweet = TweetTokenizer().tokenize(text)
tokens_whitespace = WhitespaceTokenizer().tokenize(text)
tokens_wordpunct = WordPunctTokenizer().tokenize(text)

print("BlanklineTokenizer:", tokens_blankline)
print("LegalitySyllableTokenizer:", tokens_legality_syllable)
print("LineTokenizer:", tokens_line)
print("MWETokenizer:", tokens_mwe)
print("NLTKWordTokenizer:", tokens_nltk_word)
print("PunktSentenceTokenizer:", tokens_punkt)
print("RegexpTokenizer:", tokens_regex)
print("ReppTokenizer:", tokens_repp)
print("SExprTokenizer:", tokens_sexpr)
print("SpaceTokenizer:", tokens_space)
print("StanfordSegmenter:", tokens_stanford)
print("SyllableTokenizer:", tokens_syllable)
print("TabTokenizer:", tokens_tab)
print("TextTilingTokenizer:", tokens_tiling)
print("ToktokTokenizer:", tokens_toktok)
print("TreebankWordDetokenizer:", tokens_treebank_detokenize)
print("TreebankWordTokenizer:", tokens_treebank)
print("TweetTokenizer:", tokens_tweet)
print("WhitespaceTokenizer:", tokens_whitespace)
print("WordPunctTokenizer:", tokens_wordpunct)
```

```python
#  

class MyTokenizer:
    def __init__(self):
        self.delimiters = [' ', '.', '?', ',', '!', ':', ';', '{', '}', '(', ')', '[', ']', '\t', '\n']

    def tokenize(self, text):
        tokens = []
        tokenizerr = ''
        for char in text:
            if char in self.delimiters:
                if tokenizerr:
                    tokens.append(tokenizerr)
                    tokenizerr = ''
            else:
                tokenizerr += char
        if tokenizerr:
            tokens.append(tokenizerr)
        return tokens

with open("C:\\Users\\User\\Downloads\\The_Ones_Who_Walk_Away_from_Omelas.txt", 'r', encoding='utf-8') as f:
    text = f.read()


token = MyTokenizer()
instrument = text
tokens = token.tokenize(instrument)
print(tokens)
```
