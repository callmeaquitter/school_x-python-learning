```python
class MixinDestruction:
    def destruct(self):
        print('I am pro in self-destruction')
class MixinWalking:
    def walk(self):
        print('I walk ')
class MixinSmoking:
    def smoke(self):
        print('I smoke three packs a day')
class MixinBuildable:
    def build(self):
        print('I can build you "Мачо Дача Хата Хаус"')
class MixinDustable:
    def dust(self):
        print("I am your allergy's cause")
class MixinFallable:
    def fall(self):
        print("I fall down every second")
class MixinChinesable:
    def quality(self):
        print('I was made in China')
class MixinSportable:
    def sport(self):
        print('Sorry, I cannot see you, I am in gym')


class BaseCharacter:
    def __init__(
        self, 
        name, 
        age, 
        gender
    ):
        self.name = name
        self.age = age
        self.gender = gender



class MixinFunkoPop(
    MixinDustable,
    MixinFallable,
    MixinChinesable
):
    pass

class MixinPlayable(
    MixinWalking,
    MixinBuildable,
    MixinSmoking
):
    pass

class Charlie(
    BaseCharacter, 
    MixinPlayable,
    MixinSportable
):
    def __init__(
        self, 
        name,
        age,
        gender
    ):
        super().__init__(name, age, gender)

class Steve(
    BaseCharacter,
    MixinDestruction,
    MixinWalking,
    MixinFunkoPop
):
    def __init__(
        self, 
        name, 
        age, 
        gender
    ):
        super().__init__(name, age, gender)

class Jimmy(
    BaseCharacter, 
    MixinFunkoPop,
    MixinDestruction,
):
    def __init__(
        self,
        name,
        age,
        gender
    ):
        super().__init__(name, age, gender)

class Jessi(
    BaseCharacter,
    MixinPlayable,
):
    def __init__(
        self,
        name,
        age,
        gender
    ):
        super().__init__(name, age, gender)
```
